```python
# LOAD OTHER JSON's
# load list of annotated (unique) bounding boxes in the entire dataset
with open('./caltech_bboxes_20200316.json', 'r') as file:  # Replace with your file name if different
    bbox_meta = json.load(file)

bboxes_images_df = pd.DataFrame(bbox_meta['images']) # to get the location of each annotated image, needed for train-test split

# load list suggested datasplit
with open('./CaltechCameraTrapsSplits_v0.json', 'r') as file:  # Replace with your file name if different
    data_split_meta = json.load(file)


train_locations = data_split_meta['splits']['train']
val_locations = data_split_meta['splits']['val']
```

```python
bboxes_images_df.head(3)
```

```python
len(bboxes_images_df)
```

```python
# 2.1 key (image_id) => LOCATION from bbox_meta['images']
image_id_to_location_df = bboxes_images_df[['id', 'location']].copy()
image_id_to_location_df.set_index('id', inplace = True)
image_id_to_location_df.head()
```

```python
len(image_id_to_location_df)
```

```python
# get original image file from filename
tyler_json_df['image_id'] = tyler_json_df['filename'].apply(
    lambda x: x.split('_')[0] if isinstance(x, str) else x)
tyler_json_df.head()
```

```python
tyler_json_df['location'] = tyler_json_df['image_id'].map(image_id_to_location_df['location'])
```

```python
tyler_json_df.head()
```

```python
tyler_json_df['split_type'] = tyler_json_df['location'].apply(
    lambda x: 'train' if x in train_locations else 'val' if x in val_locations else 'cis-trans')
```

```python
tyler_json_df.head()
```

```python
x = 5
tyler_json_df[tyler_json_df['category_id'] == x]['label'].iloc[0]
```

```python
# 3. show summary distribution per category 
new_ds_summary = []
new_ds_summary = pd.DataFrame({
    'category_id': np.unique(tyler_json_df['category_id']).tolist(),
    })
new_ds_summary['category_label'] = new_ds_summary['category_id'].apply(
        lambda x: tyler_json_df[tyler_json_df['category_id'] == x]['label'].iloc[0])

new_ds_summary['num_train'] = new_ds_summary['category_id'].apply( 
    lambda x: len(tyler_json_df[(tyler_json_df['category_id'] == x) & (tyler_json_df['split_type'] == 'train')]))

new_ds_summary['num_val'] = new_ds_summary['category_id'].apply( 
    lambda x: len(tyler_json_df[(tyler_json_df['category_id'] == x) & (tyler_json_df['split_type'] == 'val')]))

new_ds_summary['total'] = new_ds_summary['num_train'] + new_ds_summary['num_val']



new_ds_summary
```
